```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
h = 0 #altitude
Ms = 50 #rocket shell mass
g = 9.81
rho = 1.091 #air density
r = 0.5 # cross sectional radius of rocket
A = np.pi*r**2 
Ve = 325 # exhaust speed
Cd = 0.15
Mpo = 100
```

```python
def F(U,t):
    """
    U should be an array of [height,velocity]
    """
    Mp = max([Mpo - 20*t,0])
    dMp_dt = 20 if t < 5 else 0
    Mtotal = Mp + Ms
    
    v = U[1]
    
    Fh = v
    Fv = (-Mtotal*g + dMp_dt*Ve - 0.5*rho*v*abs(v)*A*Cd)/(Mtotal)
    return np.array([Fh,Fv])
```

```python
dt = 0.1
tMax = 40
t_values = np.arange(0,tMax,dt)
N = len(t_values)
u = np.empty((N,2))
```

```python
u[0] = np.array([0,0])
for i in range(0,N-1):
    t = t_values[i]
    u[i+1] = u[i] + dt*F(u[i],t)
```

```python
h,v = np.hsplit(u,2)
plt.figure(figsize=(8,4))   #set plot size
#plt.ylim(40,160)             #y-axis plot limits
plt.tick_params(axis='both', labelsize=14) #increase font size for ticks
plt.xlabel('t', fontsize=14) #x label
plt.ylabel('h', fontsize=14) #y label
plt.plot(t_values,h, 'k-');

plt.figure(figsize=(8,4))   #set plot size
#plt.ylim(40,160)             #y-axis plot limits
plt.tick_params(axis='both', labelsize=14) #increase font size for ticks
plt.xlabel('t', fontsize=14) #x label
plt.ylabel('v', fontsize=14) #y label
plt.plot(t_values,v, 'k-');
```
